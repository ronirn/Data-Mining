```python
import pandas as pd
from scipy import stats

print("Tipe data sebelum konversi:")
print(update_data['wheel-base'].dtype)
print(update_data['price'].dtype)

# Konversi kolom 'wheel-base' dan 'price' ke tipe numerik, dan ubah nilai non-numerik menjadi NaN
update_data['wheel-base'] = pd.to_numeric(update_data['wheel-base'], errors='coerce')
update_data['price'] = pd.to_numeric(update_data['price'], errors='coerce')

# Cek tipe data setelah konversi
print("Tipe data setelah konversi:")
print(update_data['wheel-base'].dtype)
print(update_data['price'].dtype)

# Hapus baris yang mengandung NaN di kolom 'wheel-base' dan 'price'
update_data = update_data.dropna(subset=['wheel-base', 'price'])

# Cek apakah ada NaN yang tersisa
print("Jumlah missing values setelah pembersihan:")
print(update_data[['wheel-base', 'price']].isna().sum())

# Hitung koefisien korelasi Pearson
pearson_coef, p_value = stats.pearsonr(update_data['wheel-base'], update_data['price'])

# Tampilkan hasil
print('Korelasi Pearson Koefisien adalah', pearson_coef, 'dengan p-value', p_value)
```
